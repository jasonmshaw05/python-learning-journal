```python
class HashTable:
  def __init__(self, size):
    self.array = [None] * size
    self.size = size

  def hashFunction(self, key):
      hashcode = 0
      for letters in key.lower():
        #using ordinal numbers seemed like the easiest way to do it
        hashcode += ord(letters)
      hashcode = (hashcode) % self.size
      return hashcode

  def insert(self, key, value):
    hashcode = self.hashFunction(key)
    for pairs in range(self.size):
      if self.array[hashcode] == None:
        self.array[hashcode] = (key, value)
        return
      hashcode = (hashcode + 1) % self.size
    print("There was nowhere to place the pair")

  def remove(self, key):
    hashcode = self.hashFunction(key)
    for pairs in range(self.size):
      if self.array[hashcode] != None and (self.array[hashcode])[0] == key:
        self.array[hashcode] = None
        return
      hashcode = (hashcode + 1) % self.size
    print("There was no pairing with that key")

  def lookup(self, key):
    hashcode = self.hashFunction(key)
    for pairs in range(self.size):
      if self.array[hashcode] != None and (self.array[hashcode])[0] == key:
        return self.array[hashcode]
      hashcode = (hashcode + 1) % self.size
    print("There was no pairing with that key")
```

```python
class Node:
  def __init__(self, key, data):
    self.key = key
    self.data = [data]
    self.left = None
    self.right = None

class BinaryTree:
  def __init__(self):
    self.head = None

  def lookup(self, key):
    nodeAt = self.head
    while nodeAt != None:
      if (nodeAt.key) > key:
        nodeAt = nodeAt.left
      elif nodeAt.key < key:
        nodeAt = nodeAt.right
      elif nodeAt.key == key:
        return nodeAt
    return -1

  def insert(self, key, data):
    if self.head == None:
      self.head = Node(key, data)
      return
    nodeAt = self.head
    while nodeAt != None:
      if nodeAt.key > key:
        if nodeAt.left == None:
          nodeAt.left = Node(key, data)
          return
        else:
          nodeAt = nodeAt.left
      elif nodeAt.key < key:
        if nodeAt.right == None:
          nodeAt.right = Node(key, data)
          return
        else:
          nodeAt = nodeAt.right
      elif nodeAt.key == key:
        nodeAt.data.append(data)
        return

  def remove(self, key):
    nodeWas = None
    nodeAt = self.head
    while nodeAt != None:
      if nodeAt.key > key:
        nodeWas = nodeAt
        nodeAt = nodeAt.left
      elif nodeAt.key < key:
        nodeWas = nodeAt
        nodeAt = nodeAt.right
      elif nodeAt.key == key:
        #no children
        if nodeAt.left == None and nodeAt.right == None:
          if nodeWas == None:
            self.head = None
          elif nodeWas.left == nodeAt:
            nodeWas.left = None
          else:
            nodeWas.right = None
          return
        #one child, left side
        elif nodeAt.right == None:
          if nodeWas == None:
            self.head = nodeAt.left
          elif nodeWas.left == nodeAt:
            nodeWas.left = nodeAt.left
          else:
            nodeWas.right = nodeAt.left
          return
        #one child, right side
        elif nodeAt.left == None:
          if nodeWas == None:
            self.head = nodeAt.right
          elif nodeWas.left == nodeAt:
            nodeWas.left = nodeAt.right
          else:
            nodeWas.right = nodeAt.right
          return
        #two children
        else:
          if nodeAt.right.left == None:
            if nodeWas == None:
              nodeAt.right.left = nodeAt.left
              self.head = nodeAt.right
            else:
              nodeAt.right.left = nodeAt.left
              nodeWas.right = nodeAt.right
          else:
            nodetoReplace = nodeAt
            nodeAt = nodeAt.right
            while nodeAt.left.left != None:
              nodeAt = nodeAt.left
            nodetoReplace.key = nodeAt.left.key
            nodetoReplace.data = nodeAt.left.data
            if nodeAt.left.right != None:
              nodeAt.left = nodeAt.left.right
            else:
              nodeAt.left = None
          return

  def IOTraversal(self, nodeAt):
    if nodeAt == None:
      return
    self.IOTraversal(nodeAt.left)
    print(nodeAt.key)
    self.IOTraversal(nodeAt.right)
```

```python
class Animal:
  def __init__(self, name, species, careLVL):
    self.name = name
    self.species = species
    self.careLVL = careLVL


def returnBasicCare(AnimalTree):
  animalList = ""
  if AnimalTree.lookup(3) == -1:
    return "There were no animals of that care level!"
  for animals in (AnimalTree.lookup(3)).data:
    animalList += animals.name + ", "
  return animalList[0:len(animalList) - 2]

def returnAdvancedCare(AnimalTree):
  animalList = ""
  if AnimalTree.lookup(7) == -1:
    return "There were no animals of that care level!"
  for animals in (AnimalTree.lookup(7)).data:
    animalList += animals.name + ", "
  return animalList[0:len(animalList) - 2]

def returnIntensiveCare(AnimalTree):
  animalList = ""
  if AnimalTree.lookup(10) == -1:
    return "There were no animals of that care level!"
  for animals in (AnimalTree.lookup(10)).data:
    animalList += animals.name + ", "
  return animalList[0:len(animalList) - 2]

def increaseCare(AnimalTree, nodeAt):
  if nodeAt == None:
    return
  increaseCare(AnimalTree, nodeAt.left)
  nodeAt.key += 1
  increaseCare(AnimalTree, nodeAt.right)
```
